```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('profiledata.csv')
```

```python
df.plot()
```

```python
df.plot(kind='scatter',x='media_count',y='follower_count')
```

```python
data = df[(df.follower_count < 100000)]
sns.set(color_codes=True)
```

```python
data = data[(data.following_count > 1) & (data.follower_count > 1) & (data.likes_average_count > 1) & ((df.is_private) == False)]
following = np.log(data.following_count)
follower = np.log(data.follower_count)
likes = np.log(data.likes_average_count)
```

```python
data["following_log"]=following
data["follower_log"]=follower
data["likes_log"]=likes
```

```python
sns.jointplot(x="follower_log", y="likes_log", data=data, kind="reg", joint_kws={'line_kws':{'color':'cyan'}})
slope, intercept, r_value,  p_value, std_err = linregress(data.follower_log, data.likes_log)
print("r-squared: %f" % r_value**2)
print("slope: %f    intercept: %f     p_value: %f" % (slope, intercept, p_value))
```
